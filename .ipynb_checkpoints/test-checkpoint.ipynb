```python
import pandas as pd
import fbprophet
from fbprophet import Prophet
```

```python
df=pd.read_csv('prophet_example.txt')
df.head()
```

```python
df.describe()
```

```python
m=Prophet()
m.fit(df)
```

```python
future=m.make_future_dataframe(periods=365)
```

```python
type(future)
```

```python
future.describe()
```

```python
future.head()
```

```python
forecast = m.predict(future)
forecast.tail()
```

```python
fig1 = m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```

```python
from fbprophet.plot import plot_plotly as plty
#import plotly.offline as py
plty.offline.init_notebook_mode()

fig = plot_plotly(m, forecast)  # This returns a plotly Figure
plty.iplot(fig)
```
